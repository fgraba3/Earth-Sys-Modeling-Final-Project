#### This file includes.csv files downloaded from weather.gov. The data contains air temperature profiles from Burns, Oregon. We are using this data as they are inputs into the following equations listen in Hostetler & Bartlein Similation of Lake Evaporation (1990):
- Long wave radiation from the atmosphere (20)
- ambient vapor pressure (28)
- Bowen Ration (32)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data =  pd.read_csv('AirTemp_Burns18-23.csv', delimiter=',')
years = data['YEAR']
months = data.columns[1:]

all_months = []
temps = []

for year in years:
    for month in months:
        all_months.append(f"{month}-{year}")
        temps.append(data.loc[data['YEAR'] == year, month].values[0])

plt.figure(figsize=(16, 6))
plt.plot(all_months, temps, marker='o', linestyle='-', color='b')
plt.xticks(rotation=90, fontsize=8)
plt.xlabel('Time (Month-Year)')
plt.ylabel('Temperature (F)')
plt.title('Air Temp in Burns, OR (2018-2023)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
